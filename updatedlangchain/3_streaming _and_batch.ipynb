```python
from langchain_ollama import ChatOllama
model = ChatOllama(model="gemma3:4b")
```

```python
# model.invoke("Write me a 200 words paragraph on Artificial Intelligence")
```

```python
responses = model.stream("Write me a 200 words paragraph on Artificial Intelligence")
```

```python
for chunk in responses:
    print(chunk.text, end = "", flush=True)
```

```python
for chunk in model.stream("Why do parrot have colourful feathers?"):
    print(chunk.text, end="", flush=True)
```
