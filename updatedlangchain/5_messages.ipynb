# Message Types in LangChain

In this notebook, we will explore the different types of messages in LangChain and how they are used to represent conversations with LLMs. Unlike simple string inputs, LangChain uses structured message objects to represent inputs and outputs, which is especially important for chat models.

```python
import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Ensure the API keys are set in the environment
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

## 1. Initialize the Chat Model

First, we need to initialize our `model` variable. The error `Could not find name 'model'` occurred because this notebook runs in a separate kernel/session, and the `model` variable hadn't been defined in this file yet.

```python
from langchain.chat_models import init_chat_model

# Initialize your preferred model (e.g., Google Gemini, Groq, or OpenAI)
# Let's use Gemini 2.5 Flash Lite as used in previous notebooks
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
model
```

## 2. Understanding LangChain Messages

LangChain has three primary message classes that represent different roles in a chat conversation:

1. **`SystemMessage`**: Used to guide the model's behavior, tone, or perspective (e.g., telling it to act like a specific persona or restrict its answers).
2. **`HumanMessage`**: Represents messages sent by the user.
3. **`AIMessage`**: Represents responses generated by the AI model.

Let's import these classes.

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
```

## 3. Creating and Invoking Messages

Let's pass a list of messages containing a system instruction and a user question to the `model`.

```python
messages = [
    SystemMessage(content="You are poetry expert"),
    HumanMessage(content="Write a poem on artificial intelligence?")
]

response = model.invoke(messages)
print("AI Response Object:", response)
print("\nAI Text Content:", response.content)
```

```python
system_msg = SystemMessage("You are a helpful codeing assistant.")

messages = [
    system_msg,
    HumanMessage("How do I create a REST API?")
]

response = model.invoke(messages)
print(response.content)
```

```python
### Detail info to the llm through system message

from langchain.messages import HumanMessage, SystemMessage

system_msg=("""
You are a senior Python developer with expertise in web framework.
Always provide code examples and explain your reasoning.
Be concise but through in your explanations.
""")

messages = [
    system_msg,
    HumanMessage('How do I create a REST API?')
]

response = model.invoke(messages)
print(response.content)
```

## 4. Representing Conversation History

To build conversational/chat applications, we feed the model the entire history of `HumanMessage` and `AIMessage` exchanges, allowing it to remember context.

```python
chat_history = [
    SystemMessage(content="You are a pirate captain. Answer all questions with pirate slang."),
    HumanMessage(content="Hi, what is your name?"),
    AIMessage(content="Ahoy! I be Captain Chat, ruler of the digital seas!"),
    HumanMessage(content="Where is the gold hidden?")
]

response = model.invoke(chat_history)
print(response.content)
```
