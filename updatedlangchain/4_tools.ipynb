```python
import os
from dotenv import load_dotenv
load_dotenv()


os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

models = ChatGroq(model="qwen/qwen3-32b")

resp=models.invoke("Why sky blue?")
resp
```

```python
from langchain.tools import tool

@tool
def get_weather(location:str)->str:
    """Get the weather at a location"""
    return f"It/s sunny in {location}"    
```

```python
model_with_tools = models.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("What is the weather like in boston?")
print(response)

for tool_call in response.tool_calls:
    print(f"Tools : {tool_call['name']}")
    print(f"Args : {tool_call['args']}")
```
